# WeatherPy

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import random
import csv
import pprint
import openweathermapy.core as owm

# Import API key
api_key = "f82e51275cfca026c27fe0b31e80d5fa"

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs, cities, and other variables
cities = []
lat_lngs = []
temp = []
humidity = []
cloudiness = []
wind_speed = []
lat = []
lon = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
query_url = f"{url}appid={api_key}&units={units}&q="

cities_df = pd.DataFrame()
cities_df['City'] = cities

for index, row in cities_df.iterrows():
    try:
        city_name = row['City']
        print(city_name)
        print(index)
        response = requests.get(query_url + city_name).json()
        temp.append(response['main']['temp'])
        lat.append(response['coord']['lat'])
        lon.append(response['coord']['lon'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        wind_speed.append(response['wind']['speed'])    
        print(f"Processing record | {city_name}")
        print(query_url + city_name)
    except KeyError:
        print(f"Weather for {city_name} not found.")
        temp.append("Null")
        lat.append("Null")
        lon.append("Null")
        humidity.append("Null")
        cloudiness.append("Null")
        wind_speed.append("Null")
        continue

```

### Convert Raw Data to DataFrame

```python
# Create dataframe with variables of interest
weatherpy_df = pd.DataFrame()
weatherpy_df["City"] = cities
weatherpy_df["Temperature (F)"] = temp
weatherpy_df["Latitude"] = lat
weatherpy_df["Longitude"] = lon
weatherpy_df["Humidity (%)"] = humidity
weatherpy_df["Cloudiness (%)"] = cloudiness
weatherpy_df["Wind Speed (MPH)"] = wind_speed

# Drop rows with null values / city not found
weatherpy_df_final = weatherpy_df[weatherpy_df != "Null"]
weatherpy_df_final.dropna(axis=0, how='any',inplace=True)
weatherpy_df_final.head()
```

```python
weatherpy_df_final.to_csv(r'CitiesData_NickBredahl.csv')
```
